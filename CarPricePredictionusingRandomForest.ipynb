# **Car Price Prediction using RANDOM FOREST ALGORITHM**

***Importing essential Libraries***

```python
import pandas as pd
```

***Load Dataset from Local Directory***

```python
from google.colab import files
uploaded = files.upload()
```

***Load Dataset***

```python
dataset = pd.read_csv('CarPrice.csv')
dataset = dataset.drop(['car_ID'],axis=1)
dataset
```

***Summarize Dataset***

```python
print(dataset.shape)
print(dataset.head(5))
```

***Splitting dataset into X and Y***

*This X contains Both Numerical and Text Data*

```python
Xdata=dataset.drop('price',axis='columns') #dropping price column
numericalCols=Xdata.select_dtypes(exclude=['object']).columns #excluding datatypes of object datatype
X=Xdata[numericalCols] #includes all numerical based data
X
```

```python
Y = dataset['price']
Y
```

***Scaling the Independent Variables (Features)***

```python
from sklearn.preprocessing import scale
cols = X.columns
X = pd.DataFrame(scale(X)) #rescaling of data(universal Scale)
X.columns = cols
X
```

***Splitting Dataset into Train and Test***

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0) #20% for testing, 80% for training
```

***Training Using Random Forest***

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

***Evaluating Model***

```python
ypred = model.predict(x_test)

from sklearn.metrics import r2_score
r2score = r2_score(y_test, ypred)
print("R2Score",r2score*100)
```
